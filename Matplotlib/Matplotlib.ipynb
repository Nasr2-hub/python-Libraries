```python
# Values in dict & Auto-sizing
```

```python
data={'Democratic':48,'Republican':46,'Libertarian':3}
fig=plt.figure(FigureClass=Waffle,rows=5,values=data,legend={'loc':'upper left','bbox_to_anchor':(1.1,1)})
plt.show()
```

```python
# Title,Legend,colors,Background Color,Block Color,Direction & Style
```

```python
data={'Democratic':48,'Republican':46,'Libertarian':3}
fig=plt.figure(
    FigureClass=Waffle,
    rows=5
    ,values=data,
    colors=["blue","red","yellow"],
    title={'label':'Vote Percentage in US Presidential Election','loc':'left'},
    labels=[f"{k}({v}%)"for k,v in data.items()],
    legend={'loc':'lower left','bbox_to_anchor':(0,-0.4),'ncol':len(data),'framealpha':0},
    starting_location='NW',
    block_arranging_style='snake'
)
fig.set_facecolor('#EEEEEE')
plt.show()
```

```python
# Plot With Icons-Pictogram Chart
```

```python
import matplotlib.pyplot as plt
from pywaffle import Waffle
data={'Democratic':48,'Republican':46,'Libertarian':3}
fig=plt.figure(
    FigureClass=Waffle,
    rows=5,
    values=data,
    colors=["blue","red","yellow"],
    legend={'loc':'upper left','bbox_to_anchor':(1,1)},
    icons='child',
    font_size=12,
    icon_legend=True
)
plt.show() 
```

```python
from pywaffle import Waffle
import matplotlib.pyplot as plt

data = {'Democratic': 48, 'Republican': 46, 'Libertarian': 3}

fig = plt.figure(
    FigureClass=Waffle,
    rows=5,
    values=data,
    colors=["blue", "red", "yellow"],
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)},
    icons='child',
    font_size=12,
    icon_legend=True
)

plt.show()
```
