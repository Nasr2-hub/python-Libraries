```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
car_mileage=np.array([10,15,20,25,30]).reshape(-1,1)
car_price=np.array([500000,400000,300000,200000,150000])
```

```python
model=LinearRegression()
model.fit(car_mileage,car_price)
```

```python
predicted_price=model.predict([[20]])
print(predicted_price)
```

```python
print(f"slope((coefficient):{model.coef_[0]}","predicted_price")
print(f"intercept:{model.intercept_}")
```

```python
predicted_price=model.predict(car_mileage)
plt.scatter(car_mileage,car_price,color='red',label='Actual Prices')
plt.plot(car_mileage,predicted_price,color='blue',label='Regression Line')
plt.xlabel('car_mileage(per litre)')
plt.ylabel('car_price(dollars)')
plt.title('Linear Regression:car_mileage vs car_prices')
plt.legend()
plt.show()
            
```
