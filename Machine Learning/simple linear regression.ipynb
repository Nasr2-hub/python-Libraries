```python
# example 1
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
area=np.array([1500,2000,2500,3000,3500]).reshape(-1,1)
price=np.array([300000,400000,500000,600000,700000])
```

```python
model=LinearRegression()
model.fit(area,price)
```

```python
print(f"slope:{model.coef_[0]},intercept:{model.intercept_},predicted_Price")
```

```python
predicted_price=model.predict([[3000]])
print(predicted_price)
```

```python
predicted_price=model.predict(area)
plt.scatter(area,price,color='blue',label='Actual prices')
plt.plot(area,predicted_price,color='red',label='Regression Line')
plt.xlabel('Area(square feet)')
plt.ylabel('price(dollars)')
plt.title('Simple Linear Regression:Area vs Price')
plt.legend()
plt.show()
```

```python
# example 2
```

```python
engine_size=np.array([1000,1500,2000,2500,3000]).reshape(-1,1)
car_price=np.array([20000,25000,30000,35000,40000])
```

```python
model=LinearRegression()
model.fit(engine_size,car_price)
```

```python
predicted_price=model.predict([[2500]])
print(predicted_price)
```

```python
print(f"slope((coefficient):{model.coef_[0]}","predicted_Price")
print(f"intercept:{model.intercept_}")
```

```python
predicted_price=model.predict(engine_size)
plt.scatter(engine_size,car_price,color='blue',label='Actual prices')
plt.plot(engine_size,predicted_price,color='red',label='Regression Line')
plt.xlabel('(engine_size(cubic cm)')
plt.ylabel('car_price(dollars)')
plt.title('Simple Linear Regression:Engine_size vs car_Price')
plt.legend()
plt.show()
```
