```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("C:/Users/User/Downloads/carprices.csv")
data
```

```python
dummies=pd.get_dummies(data["Car Model"])
dummies
```

```python
merge=pd.concat([data,dummies],axis=1)
merge
```

```python
merge=merge.drop(["Car Model"],axis=1)
merge
```

```python
y=merge["Sell Price($)"]
x=merge.drop(["Sell Price($)"],axis=1)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
```

```python
model.predict([[45000,4,0,0,1]])
```

```python
model.predict([[86000,7,0,1,0]])
```

```python
y_pred=model.predict(x)
y_pred
```

```python
print(x.shape)
print(y.shape)
print(y_pred.shape)
```

```python
from sklearn.metrics import r2_score
```

```python
r2=r2_score(y,y_pred)
r2
```
