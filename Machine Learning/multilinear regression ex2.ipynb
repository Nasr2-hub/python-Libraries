```python
import pandas as pd
import numpy as np
```

```python
data={
    'engine_size':[1000,1500,2000,2500,3000,3500,4000,4500,5000,5500,6000,6500,7000,7500,8000],
    'horsepower':[90,110,130,150,170,190,210,230,250,270,290,310,330,350,370],
    'weight':[1000,1200,1400,1600,1800,2000,2200,2400,2600,2800,3000,3200,3400,3600,3800],
    'fuel_efficiency':[40,38,36,34,32,30,28,26,24,22,20,18,16,14,12]
}
data
```

```python
df=pd.DataFrame(data)
df
```

```python
from sklearn.linear_model import LinearRegression
x=df[["engine_size","horsepower","weight"]]
y=df[["fuel_efficiency"]]
model=LinearRegression()
model.fit(x,y)
```

```python
model.predict([[1500,110,1200]])
```
