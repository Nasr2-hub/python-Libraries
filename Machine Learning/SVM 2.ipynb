```python
import pandas as pd
```

```python
from sklearn.datasets import load_digits
digits=load_digits()
dir(digits)
```

```python
data=pd.DataFrame(digits.data,columns=digits.feature_names)
data
```

```python
data["targets"]=digits.target
data
```

```python
digits.target_names
```

```python
data['flower_name']=data.targets.apply(lambda x:digits.target_names[x])
data
```

```python
# Assuming `data` is your DataFrame
y = data['targets']  # Replace 'target' with the actual target column name
x = data.drop(columns=['targets'])  # Replace 'target' with the target column name
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
xtrain,xtest,ytrain,ytest=train_test_split(x,y)
```

```python
model=SVC()
model.fit(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
model.predict(xtest)
```

```python
ytest
```

```python
svclassifier=SVC(kernel="linear")
svclassifier.fit(xtrain,ytrain)
svclassifier.score(xtest,ytest)
```

```python
svclassifier_2=SVC(kernel='poly',degree=8)
svclassifier_2.fit(xtrain,ytrain)
svclassifier_2.score(xtest,ytest)
```

```python
Gau_fier=SVC(kernel='rbf',random_state=0)
Gau_fier.fit(xtrain,ytrain)
Gau_fier.score(xtest,ytest)
```

```python
sig_fier=SVC(kernel='sigmoid')
sig_fier.fit(xtrain,ytrain)
sig_fier.score(xtest,ytest)
```
